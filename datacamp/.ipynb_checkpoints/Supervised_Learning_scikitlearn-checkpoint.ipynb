```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets
from sklearn.model_selection import train_test_split
plt.style.use('ggplot')
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

```python
type(iris)
```

```python
print(iris.keys())
```

```python
print(iris.DESCR)
```

```python
X, X_new = train_test_split(iris.data, test_size = 0.1, shuffle=False)
y, y_new = train_test_split(iris.target, test_size = 0.1, shuffle=False)
df = pd.DataFrame(X, columns=iris.feature_names)
df_new = pd.DataFrame(X_new, columns=iris.feature_names)
```

```python
_ = pd.scatter_matrix(df, c = y, figsize = (10,10), s = 50, marker = 'D')
_ = pd.scatter_matrix(df_new, c = y_new, figsize = (10,10), s = 50, marker = 'D')
```

```python
plt.scatter(df['petal length (cm)'],df['petal width (cm)'], c = y)
plt.xlabel('petal width (cm)')
plt.ylabel('petal length (cm)')
plt.title('petal width vs length')
plt.legend(iris.target_names)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=6)
```

```python
knn.fit(iris['data'],iris['target'])
```

```python
prediction = knn.predict(X_new)
```

```python
X_new.shape
```

```python
print('prediction {}'.format(prediction))
```
