```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/facebookincubator/prophet/master/examples/example_wp_peyton_manning.csv')
```

```python
df['y'] = np.log(df['y'])
df.tail()
```

```python
p = Prophet()
p.fit(df)
```

```python
future = p.make_future_dataframe(periods=365)
future.tail()
```

```python
df_pred = p.predict(future)
df_pred[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```
